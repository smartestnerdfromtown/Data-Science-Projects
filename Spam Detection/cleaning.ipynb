```python
import pandas as pd
```

```python
df = pd.read_csv(
    filepath_or_buffer="mail_data.csv"
)
```

```python
df.head()
```

```python
df.head(1)["Message"].values
```

```python
def text_processing(df: pd.DataFrame, column: str) -> pd.DataFrame:
    df_processed = (
        df[column]
        .str.lower()
        .str.replace(r"http\S+|www\S+", "<URL>", regex=True)
        .str.replace(r"\d+", "<NUM>", regex=True)
        .str.replace(r"[^\w\s]", "", regex=True)
    )

    return df_processed
```

```python
df_processed = text_processing(df=df, column="Message")
```
